```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
x= np.linspace(-10,10,1000)
y= np.sin(x)
z= np.cos(x)
```

```python
plt.style.use ('seaborn')
plt.plot(x,y, label = 'sin')
plt.plot(x,z, label = 'cos')
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.legend()# gives the information of plot 
```

```python
fig = plt.figure(figsize =(8, 4))
#ax1 = fig.add_axes([0.1,0.8,0.2,0.1])
ax2 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1 = fig.add_axes([0.1,0.7,0.2,0.1])
ax3 = fig.add_axes([0.7,0.7,0.2,0.1])
ax1.plot(x,y)
ax2.plot(x,z,color = 'green')
ax3.plot(x,z,color = 'k')
plt.show()
```
